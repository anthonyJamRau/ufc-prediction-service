```python
for country_code in country_code_map.keys():

    request_url = f"https://www.ufc.com/athletes/all?filters[0]=location:{country_code}"
    response = requests.get(request_url,verify=False)
    location_page_soup = bs4.BeautifulSoup(response.text)
    fighters = [fighter for fighter in location_page_soup.find_all('li', class_='l-flex__item') if fighter.find('div', class_='node node--type-athlete node--view-mode-all-athletes-result ds-1col clearfix')]

    for fighter in fighters:
        name_tag = fighter.find('span',class_='c-listing-athlete__name')
        name = name_tag.get_text(strip=True)
        image_tag = fighter.find('img',class_='image-style-teaser')
        image_url = image_tag['src'] if image_tag else 'NULL'
        df_store.loc[len(df_store.index)] = [name,image_url,country_code,country_code_map[country_code]]

    time.sleep(1)
```

```python
# row by row to slow requests
for index,row in df_store.iterrows():
    image_url = row['image_path']
    if image_url != 'NULL':
        response = requests.get(image_url, verify=False)
        if response.status_code == 200:
            with open(f"data/images/{row['fighter_name']}.png", 'wb') as f:
                f.write(response.content)
        else:
            logging.warning(f"{image_url} not accessed. Code {response.status_code}")
        time.sleep(1)
    else:
        continue
```
